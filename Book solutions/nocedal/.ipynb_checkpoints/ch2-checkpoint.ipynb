```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    return((x-1.2)**2+(y-4)**2)
```

```python
xx = np.linspace(-20,20,100)
yy = np.linspace(-20,20,100)
Z = np.zeros((len(xx),len(yy)))

          
for i in range(len(xx)):
    for j in range(len(yy)):
        Z[i,j] = 4*xx[i]**2 + yy[j]**2
```

```python
fig, ax = plt.subplots()
ax.contour(xx,yy,Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
```

The contour lines should be stretched in the $y$ direction, not the $x$ direction! Why is this happening? The indexing order is reversed.

Think of placing a cartesian coordinate system atop our matrix. Then $x$ increases from left-to-right (across the columns) and $y$ increases from top-to-bottom (down the rows). Hence, indexing order is reversed. We can fix this by using the transpose of our function:

```python
fig, ax = plt.subplots()
ax.contour(xx,yy,Z.T)
ax.set_xlabel('x')
ax.set_ylabel('y')
```

The transpose exchanges rows and columns. Now, the original matrix's rows are counted from left-to-right and columns are counted from top-to-bottom.

Most 3D plotting routines require domain data to be 2D arrays. We can use `np.meshgrid` to turn vectors into coordinate matrices. When we use this on two vectors running along our axes, we get back two matrices, one for $x$-coordinates and one for $y$-coordinates. The values in $X$ increase from left-to-right, and the values in $Y$ increase from top-to-bottom.

`meshgrid` allows us to easily vectorize functions by passing these coordinate arrays in an performing elementwise operations between them, rather than looping over the elements.

Alternately, we can use the `meshgrid` option `indexing = 'ij'` to indicate that we want to plot a function using row,col ordering rather than x,y ordering.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
#%matplotlib notebook
```

```python
def f(X,Y):
    return ((X-3)**2 + 4*(Y-3)**2)
def dfx(X,Y):
    return (2*(X-3))
def dfy(X,Y):
    return (8*(Y-3))
```

```python
xx = np.linspace(-10,10,100)
yy = np.linspace(-10,10,100)

X,Y = np.meshgrid(xx,yy)
Z = f(X,Y)

ZX = dfx(X,Y)
ZY = dfy(X,Y)

fig = plt.figure(figsize=(10,30))
ax = fig.add_subplot(3,1,1, projection='3d')
ax.plot_surface(X,Y,Z,alpha=0.8)
ax.scatter(3,3,0,s=30,c='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.azim=30
ax.elev=45

ax = fig.add_subplot(3,1,2, projection='3d')
ax.plot_surface(X,Y,ZX,alpha=0.8)
ax.scatter(3,3,0,s=30,c='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.azim=30
ax.elev=45

ax = fig.add_subplot(3,1,3, projection='3d')
ax.plot_surface(X,Y,ZY,alpha=0.8)
ax.scatter(3,3,0,s=30,c='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.azim=30
ax.elev=45
```

```python
xr = np.array([9,-4])


def sditer(xr,lr=1):
    """ Take a step of length lr in gradient descent. """
    p = np.array([dfx(xr[0],xr[1]), dfy(xr[0],xr[1])])
    if lr != 0.:
        p = p / np.linalg.norm(p)
        return(xr - lr*p)
    else:
        return(xr - 0.05*p)
        
x = np.zeros((15,2))
x[0,:] = [9,-4]

for j in range(1,15):
    x[j,:] = sditer(x[j-1,:],1)
```

```python
fig = plt.figure(figsize=(10,30))
ax = fig.add_subplot(3,1,1, projection='3d')
ax.plot_surface(X,Y,Z,alpha=0.8)
ax.scatter(3,3,0,s=30,c='g')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.azim=30
ax.elev=45

ax.scatter(x[:,0],x[:,1],f(x[:,1],x[:,1]),c='r',s=60)

plt.show()
```
